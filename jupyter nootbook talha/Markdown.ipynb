# 10- Adding code or code block

To adding a code in Markdown use backtick( ` )

To print a string use `print("Python")`

`print("Hello World")`

To adding a block of code in Markdown use triple backtick( ``` )

> This code with show color according to python language syntax

```python
x = 5+4
y = 4-2
z = x+y
print(z)
```

> This code with show color according to R language syntax

```r
x = 5+4
y = 4-2
z = x+y
print(z)
```
