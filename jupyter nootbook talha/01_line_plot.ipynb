Seaborn automatically install these libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
flowers = sns.load_dataset("iris")
flowers
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flowers)
plt.show()
```

## Adding titles

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
flowers = sns.load_dataset("iris")
flowers
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flowers)
plt.title("Flowers Plot")

plt.show()
```

## Adding limits

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
flowers = sns.load_dataset("iris")
flowers
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flowers)
plt.title("Flowers Plot")
plt.xlim(3)
plt.ylim(2)

plt.show()
```

## Set styles
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
flowers = sns.load_dataset("iris")
flowers

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flowers)
plt.title("Flowers Plot")
sns.set_style("dark")
sns.set_style(style=None, rc=None)

plt.show()
```

## Size of figure

```python
# import library
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
flowers = sns.load_dataset("iris")

# change figure
plt.figure(figsize=(10,8))

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flowers)
plt.title("Flowers Plot")

plt.show()
```
